### 声音编码

- 将波数字化

- **量化位数：**存储一个采样所用二进制数的位数
- **声道数：**同时录放的采样链个数
- **数据存储量计算公式：**

$$
数据存储量=采样频率\times量化位数\times持续秒数\times声道数\div8\ (字节)
$$

- **格式：**无压缩音频常为WAV格式

### 图像编码

- 将颜色数字化 —— 像素
- **分辨率：**水平方向上的像素数量乘以垂直方向上的像素数量（如$640\times480$）
- **位深：**存储一个像素所用二进制数的位数（红、绿、蓝三种通道，有时还有透明度通道）
- **数据存储量计算公式：**

$$
数据存储量=分辨率\times位深\div8\ (字节)
$$

- **格式：**无压缩图像常为BMP格式

### 数据压缩

- 有损压缩

  - MP3（aka. MPEG3，音频）
