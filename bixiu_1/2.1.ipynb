# 第二章 算法与程序实现 - 第一节

本节主要介绍算法的概念与描述方法，以及使用计算机编程解决问题的过程。

## 算法与算法描述

### 算法

在有限步骤内求解某一问题所使用的步骤和方法

#### 算法的特征

- 有穷性
- 确定性
- 可行性
- 有零个（隐式）或多个输入
- 有一个或多个输出

### 算法的描述

- 自然语言
  - 通俗易懂
  - 冗长，不够直观、清晰
  - 容易出现歧义（不确定性）
- 流程图
  - <img src="img/2/flow_chart.png">
  - 直观、清晰
  - 占用篇幅较大
  - 不易修改
- 伪代码
  - 能够较容易地被转换成程序设计语言

### 算法的基本控制结构

- 顺序结构
  - 自上而下，按先后顺序执行
- 分支结构（选择结构）
  - 判断条件
  - 满足条件时执行一个分支，不满足条件时执行另一个分支
- 循环结构（重复结构）
  - 根据条件重复执行某一部分操作
  - **循环体：**重复执行的部分
  - **当循环：**先判断循环条件，后执行循环体
  - **直到循环：**先执行循环体，再判断循环条件是否成立

## 编程解决问题的过程

1. 抽象与建模
   - 对问题进行深入分析，明确问题的需求
   - 已知，条件，求解
   - 用数学符号描述计算模型
2. 设计算法
   - 提炼关键步骤
   - 合理使用三种基本控制结构（顺序结构，分支结构，循环结构）
   - **符合算法三大特征**
3. 编写程序
   - 计算机执行算法的具体步骤的实现
4. 调试运行
   - 修复错误
   - 对运行结果进行分析和验证
   - 可能需要回到以上任何一步来解决问题
