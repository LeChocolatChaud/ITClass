## 顺序结构的 Python 实现

> **小复习**
>
> 顺序结构的特点：自上而下，按先后顺序执行

-   按照解决问题的顺序写出相应的语句

-   例：

    -   输入语句

        ``` python
        input(prompt=None, /) -> str
        ```

        从标准输入流（命令行）中获取字符串，不包括换行符。

        如果提供了`prompt`参数，则该参数的内容会在获取字符串前被输出到标准输出流（命令行），输出后不换行。

    -   输出语句

        ``` python
        print(*values: object, sep: str | None = ' ', end: str | None = '\n', file: SupportsWrite[str] | None = sys.stdout, flush: bool = False)
        ```

        输出`values`参数的内容。默认输出至标准输出流（命令行），可通过`file`参数指定输出流。

        可选参数：

        `file`：一个流 (stream) ，默认为`sys.stdout`。

        `sep`：多个输出内容间的插入内容，默认为一个空格。

        `end`：输出最后添加的内容，默认为换行符。

        `flush`：是否要强制刷新流的写入缓冲区。

    -   赋值语句

        详见《第二章 算法与程序实现（中）》

### 实例

## 分支结构的 Python 实现

> **小复习**
>
> 分支结构的特点：有判断条件；满足条件时执行一个分支，不满足条件时执行另一个分支

-   能根据是否满足条件来执行不同的语句

-   **实现**：if 语句

    -   单分支语句

        ``` python
        if 条件表达式:
            语句块
        ```

        > Python
        > 程序要求代码**全部使用缩进来分层**，否则将导致程序错误，无法运行。
        >
        > Python
        > 编程规范中指出：缩进最好**采用空格形式**，每一层**向右缩进4个空格**。在同一段代码中**不能混用Tab键和空格键**。

        1.  判断条件表达式

        2.  -   结果为真（`True`）
                -   执行语句块
            -   结果为假（`False`）
                -   不执行任何语句

    -   双分支语句

        ``` python
        if 条件表达式:
            语句块1
        else:
            语句块2
        ```

        1.  判断条件表达式

        2.  -   结果为真（`True`）
                -   执行语句块1
            -   结果为假（`False`）
                -   执行语句块1

    -   多分支语句

        ``` python
        if 条件表达式1:
            语句块1
        elif 条件表达式2:
            语句块2
        ······
        else:
            语句块n
        ```

        1.  判断条件表达式1

        2.  -   结果为真（`True`）
                -   执行语句块1
            -   结果为假（`False`）
                -   继续下一步

        3.  判断条件表达式2

        4.  -   结果为真（`True`）
                -   执行语句块2
            -   结果为假（`False`）
                -   继续下一步

        5.  ......

        6.  判断条件表达式n-1

        7.  -   结果为真（`True`）
                -   执行语句块n-1
            -   结果为假（`False`）
                -   继续下一步

        8.  执行语句块n

### 实例

**目标**：使用分支结构对BMI指数进行范围判定，并相应地显示 “正常” 或
“需注意” 的提示。

#### 1. 抽象与建模

将输入的BMI指数与标准范围值进行比较，并根据结果输出相应的文字描述。

-   **输入**：$bmi$（BMI指数）

-   **输出**：“正常” 或者 “需注意”

-   **模型**：

    | 描述   | 标准范围值                 |
    |--------|----------------------------|
    | 需注意 | $\mathrm{BMI}<18.5$        |
    | 正常   | $18.5\le\mathrm{BMI}\le25$ |
    | 需注意 | $\mathrm{BMI}>25$          |

#### 2. 设计算法

1.  输入BMI指数
2.  四舍五入BMI指数至小数点后1位
3.  判定BMI指数是否大于等于18.5且小于等于25。如果为真，则输出
    “正常”；如果为假，则输出 “需注意”

#### 3. 编写程序

``` python
bmi = float(input("请输入BMI指数："))
bmi = round(bmi, 1)
if bmi >= 18.5 and bmi <= 25:
    print("正常")
else:
    print("需注意")
```

[程序文件](py/compare_bmi.py)（Ctrl+左键）

-   使用`round`函数进行四舍五入操作
-   使用`if`语句进行条件判断

#### 4. 调试运行

````python
%%space branch_demo

bmi = float(input("请输入BMI指数："))
bmi = round(bmi, 1)
if bmi >= 18.5 and bmi <= 25:
    print("正常")
else:
    print("需注意")
````

## 循环结构的 Python 实现

> **小复习**
>
> 循环结构的特点：根据条件重复执行某一部分操作

-   具有规律性的重复操作

-   **实现**：

    -   `while` 语句

        ``` python
        while 条件表达式:
            语句块
        ```

        1.  判断条件表达式

        2.  -   如果为真（`True`）
                -   执行语句块
                -   跳回第一步
            -   如果为假（`False`）
                -   结束循环

    -   `for` 语句

        ``` python
        for 循环变量 in 序列:
            语句块
        ```

        1.  从序列中获取第1个元素赋值给循环变量
        2.  执行语句块
        3.  从序列中获取第2个元素赋值给循环变量
        4.  执行语句块
        5.  ......
        6.  从序列中获取第n个元素赋值给循环变量
        7.  执行语句块

        -   序列可以是`range`函数产生的列表

            ``` python
            range(stop: int, /)
            ```

            ``` python
            range(start: int, stop: int, step: int = 1, /)
            ```

            返回能产出一串整数的对象。产出的整数遵循“左闭右开原则”，从`start`（默认为0）到`stop`以`step`（默认为1）为步长递增。

### 实例

**目标**：输出多日的BMI指数

#### 1. 抽象与建模

计算某人一周的BMI指数，并与星期一一对应输出

-   **输入**：$height$（身高，单位为米），一周的体重

-   **输出**：BMI指数与对应的星期

-   **计算模型**： $$
    一周\mathrm{BMI}指数列表=\bigcup{}^7_{i=1}item_i\mathrm，item_i={weight_i\over height^2}。
    $$ 其中，$weight_i$代表一周中每日的体重，$\bigcup$为并集符号。

#### 2. 设计算法

1.  输入身高$height$
2.  输入体重$weights$
3.  读取当日体重$weight$并转换类型
4.  计算BMI指数
5.  重复执行步骤2，直到所有体重数据被读入并计算，循环结束
6.  显示对应的星期与BMI指数

#### 3. 编写程序

``` python
weekdays = ["SUN", "MON", "TUE", "WED", "THU", "FRI", "SAT"]
height = float(input("请输入身高(m): "))
weights = input("请输入一周的体重(kg)，用空格分隔：").split()

bmis = []
for weight in weights:
    bmis.append(round(float(weight)/height**2, 1))

for i in range(7):
    print(weekdays[i], "- BMI =", bmis[i])
```

[程序文件](py/week_bmi.py)（Ctrl+左键）

#### 4. 调试运行

````python
%%space loop_demo

weekdays = ["SUN", "MON", "TUE", "WED", "THU", "FRI", "SAT"]
height = float(input("请输入身高(m): "))
weights = input("请输入一周的体重(kg)，用空格分隔：").split()

bmis = []
for weight in weights:
    bmis.append(round(float(weight)/height**2, 1))

for i in range(7):
    print(weekdays[i], "- BMI =", bmis[i])
````

## 枚举法

列举出所有可能的情况，然后逐个判断有哪些符合问题所要求的条件，从而得到问题的解答。

### 一般模式

1.  确定范围
    -   **尽可能缩小枚举范围**，减少枚举次数，缩短求解时间，提高解决问题的效率
2.  验证条件

解决同一个问题，可以有多个不同的算法，不同算法的效率也可能不同。针对问题解决选择合适的算法，对程序的效率乃至问题解决的效率都有很大的影响。

### 程序实现（Python）

-   可以采用循环结构实现列举的过程
-   判断有哪些情况符合问题所要求的条件可以采用分支结构来实现

#### 实例

**目标**：显示某用户的训练课程完成情况

##### 1. 抽象与建模

-   **输入**：跑步课程中的每日训练内容和对应训练内容消耗的卡路里数，这些信息由程序自动读取数据库中存储的数据，不需要用户输入

-   **输出**：完成的项目数和未完成的项目名称

    -   **计算模型**： $$
        已完成项目数=\sum{}^n_{i=1}data_i\mathrm，data_i=\begin{cases}0\mathrm{（消耗的卡路里数=0），}\\1\mathrm{（消耗的卡路里数\ne0）；}\end{cases}$$
        $$
        未完成项目=\bigcup{}^n_{i=1}item_i\mathrm，item_i=\begin{cases}\mathrm{空字符串（消耗的卡路里数=0），}\\\mathrm{项目名称（消耗的卡路里数\ne0），}\end{cases}
        $$
        其中，n为该跑步课程共有多少项训练内容，消耗的卡路里数为大于等于零的整数，$\sum$为求和符号。

##### 2. 设计算法

**枚举法的一般模式**：

1.  确定范围
    -   用户一周的跑步课程训练内容
2.  验证条件
    -   检测某训练内容消耗的卡路里数是否为0，如不为0，则表示已完成训练，并计数；如为0，则表示该训练内容未完成，需要记录该训练内容。

**算法**：

1.  读入一周的跑步课程训练内容
2.  读入对应课程训练内容消耗的卡路里数
3.  初始设定已完成项目数$finished$为0，未完成项目名称$unfinished$为空字符串
4.  逐一列举一周的对应训练内容消耗的卡路里数
5.  如果当前枚举的卡路里数为0，则将对应的训练内容名称加入$unfinished$，否则$finished$的计数加1
6.  输出完成项数目及未完成项目名称

##### 3. 编写程序

``` python
itemlist = ["低速低强度", "变速练习", "低速低强度", "快走", "低速低强度", "坡度练习", "低速低强度"]
datalist = [400, 600, 380, 0, 420, 620, 397]
finished = 0
unfinished = ""
for i in range(7):
    if datalist[i] == 0:
        unfinished = unfinished + itemlist[i] + " "
    else:
        finished = finished + 1
print("完成了", finished, "项")
print("未完成项目:", unfinished)
```

##### 4. 调试运行
